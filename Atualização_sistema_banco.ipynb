```python
class Endereco:
    def __init__(self, rua, numero, cidade, estado, cep):
        self.rua = rua
        self.numero = numero
        self.cidade = cidade
        self.estado = estado
        self.cep = cep

class Cliente:
    def __init__(self, nome, cpf, endereco):
        self.nome = nome
        self.cpf = cpf
        self.endereco = endereco

class ContaBancaria:
    num_conta = 1

    def __init__(self, titular):
        self.agencia = "0001"
        self.numero_conta = str(ContaBancaria.num_conta)
        ContaBancaria.num_conta += 1
        self.titular = titular
        self.saldo = 0
        self.movimentacoes = []

    def deposito(self, valor):
        if valor > 0:
            self.saldo += valor
            self.movimentacoes.append(f"Depósito de R${valor:.2f}")
            return True
        else:
            return False

    def saque(self, valor):
        if valor > 0 and valor <= self.saldo:
            self.saldo -= valor
            self.movimentacoes.append(f"Saque de R${valor:.2f}")
            return True
        else:
            return False

    def extrato(self):
        print(f"Titular da conta: {self.titular.nome}")
        print(f"Agência: {self.agencia}")
        print(f"Número da conta: {self.numero_conta}")
        print(f"Saldo atual: R${self.saldo:.2f}")
        print("Histórico de movimentações:")
        for movimento in self.movimentacoes:
            print("- " + movimento)

class Banco:
    def __init__(self):
        self.clientes = []
        self.contas = []

    def cadastrar_cliente(self, nome, cpf, endereco):
        for cliente in self.clientes:
            if cliente.cpf == cpf:
                print("CPF já cadastrado.")
                return
        novo_cliente = Cliente(nome, cpf, endereco)
        self.clientes.append(novo_cliente)
        print("Cliente cadastrado com sucesso.")

    def criar_conta(self, cpf_titular):
        for cliente in self.clientes:
            if cliente.cpf == cpf_titular:
                nova_conta = ContaBancaria(cliente)
                self.contas.append(nova_conta)
                print("Conta criada com sucesso.")
                return
        print("Cliente não encontrado.")

def main():
    banco = Banco()

    while True:
        print("\nEscolha uma operação:")
        print("1 - Cadastrar cliente")
        print("2 - Criar conta")
        print("3 - Realizar depósito")
        print("4 - Realizar saque")
        print("5 - Ver extrato")
        print("6 - Sair")
        escolha = input()

        if escolha == "1":
            nome = input("Nome do cliente: ")
            cpf = input("CPF do cliente: ")
            rua = input("Rua do cliente: ")
            numero = input("Número do cliente: ")
            cidade = input("Cidade do cliente: ")
            estado = input("Estado do cliente: ")
            cep = input("CEP do cliente: ")
            endereco = Endereco(rua, numero, cidade, estado, cep)
            banco.cadastrar_cliente(nome, cpf, endereco)
        elif escolha == "2":
            cpf_titular = input("CPF do titular da conta: ")
            banco.criar_conta(cpf_titular)
        elif escolha == "3":
            num_conta = input("Número da conta: ")
            valor = float(input("Valor do depósito: "))
            for conta in banco.contas:
                if conta.numero_conta == num_conta:
                    if conta.deposito(valor):
                        print("Depósito realizado com sucesso.")
                    else:
                        print("Valor de depósito inválido.")
                    break
            else:
                print("Conta não encontrada.")
        elif escolha == "4":
            num_conta = input("Número da conta: ")
            valor = float(input("Valor do saque: "))
            for conta in banco.contas:
                if conta.numero_conta == num_conta:
                    if conta.saque(valor):
                        print("Saque realizado com sucesso.")
                    else:
                        print("Valor de saque inválido ou saldo insuficiente.")
                    break
            else:
                print("Conta não encontrada.")
        elif escolha == "5":
            num_conta = input("Número da conta: ")
            for conta in banco.contas:
                if conta.numero_conta == num_conta:
                    conta.extrato()
                    break
            else:
                print("Conta não encontrada.")
        elif escolha == "6":
            print("Saindo do sistema.")
            break
        else:
            print("Opção inválida. Por favor, escolha novamente.")

if __name__ == "__main__":
    main()
```
